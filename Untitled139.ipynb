```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()

df = pd.read_csv(cwd + '/filtered_data.csv', index_col=0)
```

```python
df
```

```python
df = df.drop(['sales_channel', 'trip_type', 'route', 'booking_origin'], axis=1)
```

```python
y = df['booking_complete']
```

```python
df = df.drop(['booking_complete'], axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test, y_train, y_test = train_test_split(df,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_train = sc.fit_transform(df_train)
df_test = sc.transform(df_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(df_train, y_train)
```

```python
y_pred  = classifier.predict(df_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred, y_test)
print(cm)
accuracy_score(y_pred, y_test)
```
